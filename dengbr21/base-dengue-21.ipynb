Substituição dos valores vazios por "Desconhecido"

Substituição dos valores das colunas contendo os valores de sintomas

```python
dengue['FEBRE'] = dengue['FEBRE'].replace({1: 'Sim', 2: 'Não'})
dengue['MIALGIA'] = dengue['MIALGIA'].replace({1: 'Sim', 2: 'Não'})
dengue['CEFALEIA'] = dengue['CEFALEIA'].replace({1: 'Sim', 2: 'Não'})
dengue['EXANTEMA'] = dengue['EXANTEMA'].replace({1: 'Sim', 2: 'Não'})
dengue['VOMITO'] = dengue['VOMITO'].replace({1: 'Sim', 2: 'Não'})
dengue['NAUSEA'] = dengue['NAUSEA'].replace({1: 'Sim', 2: 'Não'})
dengue['DOR_COSTAS'] = dengue['DOR_COSTAS'].replace({1: 'Sim', 2: 'Não'})
dengue['CONJUNTVIT'] = dengue['CONJUNTVIT'].replace({1: 'Sim', 2: 'Não'})
dengue['ARTRITE'] = dengue['ARTRITE'].replace({1: 'Sim', 2: 'Não'})
dengue['ARTRALGIA'] = dengue['ARTRALGIA'].replace({1: 'Sim', 2: 'Não'})
dengue['PETEQUIA_N'] = dengue['PETEQUIA_N'].replace({1: 'Sim', 2: 'Não'})
dengue['LEUCOPENIA'] = dengue['LEUCOPENIA'].replace({1: 'Sim', 2: 'Não'})
dengue['LACO'] = dengue['LACO'].replace({1: 'Sim', 2: 'Não'})
dengue['DOR_RETRO'] = dengue['DOR_RETRO'].replace({1: 'Sim', 2: 'Não'})
dengue['DIABETES'] = dengue['DIABETES'].replace({1: 'Sim', 2: 'Não'})
dengue['HEMATOLOG'] = dengue['HEMATOLOG'].replace({1: 'Sim', 2: 'Não'})
dengue['HEPATOPAT'] = dengue['HEPATOPAT'].replace({1: 'Sim', 2: 'Não'})
dengue['RENAL'] = dengue['RENAL'].replace({1: 'Sim', 2: 'Não'})
dengue['HIPERTENSA'] = dengue['HIPERTENSA'].replace({1: 'Sim', 2: 'Não'})
dengue['ACIDO_PEPT'] = dengue['ACIDO_PEPT'].replace({1: 'Sim', 2: 'Não'})
dengue['AUTO_IMUNE'] = dengue['AUTO_IMUNE'].replace({1: 'Sim', 2: 'Não'})

display(dengue)
```

Conversão de colunas de data para formato DateTime, com o parâmetro de "erros=coerce" para converter as datas válidas e transformar os valores inválido (definidos como 'Desconhecido' anteriormente) em Not a Time.

```python
dengue['DT_NOTIFIC'] = pd.to_datetime(dengue['DT_NOTIFIC'], errors='coerce')
dengue['DT_SIN_PRI'] = pd.to_datetime(dengue['DT_SIN_PRI'], errors='coerce')
dengue['DT_INVEST'] = pd.to_datetime(dengue['DT_INVEST'], errors='coerce')

dengue.info()
```

Pós processamento

Importar a biblioteca de gráficos e operações groupby que envolvem alguma combinação de divisão do objeto, aplicação de uma função e combinação dos resultados. Isso pode ser usado para agrupar grandes quantidades de dados e computar operações nesses grupos.

```python
import matplotlib.pyplot as plt
dengue_data_notific = dengue.groupby(dengue['DT_NOTIFIC'].dt.to_period('M')).mean(numeric_only=True)
```

Plotar o gráfico para média dos meses, com bases no período, ignorando os valores vazios no formato NaT.

```python
# Plotar o gráfico de média de Meses
plt.figure(figsize=(20, 6))
plt.plot(dengue_data_notific.index.astype(str), dengue_data_notific, marker='o', linestyle='-', color='skyblue')
plt.title('Média de Valores por Mês')
plt.xlabel('Mês e Ano')
plt.ylabel('Valor Médio')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

Ordenação das unidades por quantidade de registros para selecionarmos as 10 mais registradas.

Estudo dos sintomas registrados, com agrupamento dos valores

```python
dengue_sintomas = dengue.iloc[:, 23:45]
dengue_sintomas.head()
```

```python
dengue_sintomas.describe()
```

Concluímos, portanto que os valores únicos serão: Sim, Não e Desconhecido como definidos anteriormente no código. Vemos os números de frequência e os top valores mais registrados.

Localizando o nome "Sim" nas colunas especificas e fazendo uma soma geral dos valores localizados.

```python
sintomas_sim = (dengue_sintomas == 'Sim').sum()
sintomas_sim
```

Plotando o gráfico com base nos sintomas registrados.

```python
plt.figure(figsize=(18, 6))
plt.plot(sintomas_sim.index.astype(str), sintomas_sim, marker='o', linestyle='-', color='purple')
plt.title('Sintomas Registrados')
plt.xlabel('Sintomas')
plt.ylabel('Contagem total')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(18, 6))
plt.bar(sintomas_sim.index.astype(str), sintomas_sim, color='purple')
plt.title('Sintomas Registrados')
plt.xlabel('Sintomas')
plt.ylabel('Contagem total')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

Calcular uma porcentagem de sintomas mais registrados com base em todos

```python
sintomas_porcentagem = (dengue_sintomas == 'Sim').mean() * 100
sintomas_porcentagem
```

Plotando um gráfico de pizza

```python
maior_proporcao = sintomas_porcentagem.max()
coluna_mais_sim = sintomas_porcentagem.idxmax()

labels = sintomas_porcentagem.index
sizes = sintomas_porcentagem.values

# Configurar o gráfico de pizza
plt.figure(figsize=(20, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Porcentagem de Sintomas Registrados como "Sim"')
plt.annotate(f'Maior proporção em {coluna_mais_sim} com {maior_proporcao:.1f}%', xy=(0.5, -0.1), xytext=(0, -1.2),
             fontsize=12, ha='center', va='center', xycoords='axes fraction', textcoords='offset points')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```
