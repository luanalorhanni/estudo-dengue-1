```python
dengue_sintomas = dengue_sintomas.fillna('0')
dengue_sintomas.value_counts()
```

```python
x = dengue_sintomas #variavel preditora - entrada
y = dengue['CLASSI_FIN'][x.index] #variavel target - saida


from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3, random_state=42) 
```

```python
from sklearn.ensemble import ExtraTreesClassifier

modelo = ExtraTreesClassifier()

#Fornecer variáreis para o modelo.
modelo.fit(x_treino, y_treino)
resultado = (modelo.score(x_teste, y_teste)) * 100
print(f'A máquina registrou uma probabilidade de {resultado:.2f}% de acerto.')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC

modelos = [
    LogisticRegression(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    GradientBoostingClassifier(),
    SVC()
]

for modelo in modelos:
    modelo.fit(x_treino, y_treino)
    resultado = modelo.score(x_teste, y_teste)
    print(f'{modelo.__class__.__name__} score: {resultado}')
```

<b>Visualização baseados nos dados reais utilizados no treinamento e desempenho do modelo. </b> Aproximação daquilo que se relaciona com X e Y

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import confusion_matrix, roc_curve, auc
```

```python
x = dengue_sintomas #variavel preditora - entrada
y = dengue['CLASSI_FIN'][x.index] #variavel target - saida


from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3, random_state=42) 
```

<b>ExtraTreesClassifier</b>: é um algoritmo de aprendizado supervisionado baseado em ensembles de árvores de decisão. Ele é uma variação do Random Forest que utiliza árvores de decisão adicionais (extra trees) para aumentar a diversidade e, potencialmente, melhorar a precisão e a robustez do modelo. As previsões das árvores são combinadas para formar a previsão final. No aprendizado supervisionado, os dados de treinamento que você fornece ao algoritmo 
incluem as soluções desejadas, chamadas de rótulos.

```python
modelo = ExtraTreesClassifier(n_estimators=100, random_state=42)
modelo.fit(x_treino, y_treino)

#n_estimators = números de árvores de dcisão que serão construídas no ensemble.
#Cada árvore é construída usando um subconjunto aleatório dos dados de treinamento e dos atributos.
```

<b>Importância das Features:</b> É calculada como a redução média da impureza trazida por aquela feature em todas as árvores do ensemble. Features que resultam em maiores reduções de impureza são consideradas mais importantes.

```python
importances = modelo.feature_importances_
indices = np.argsort(importances)[::-1] #índices que ordenariam o array de importâncias invertido

# Plotar as importâncias das features
plt.figure(figsize=(10, 6))
plt.title("Importância das Features")
plt.bar(range(x.shape[1]), importances[indices], align="center")
plt.xticks(range(x.shape[1]), x.columns[indices], rotation=90)
plt.xlim([-1, x.shape[1]])
plt.show()
```

 Os métodos <b>'model.feature_importances_'</b> e <b>'np.argsort(importances)[::-1]'</b> são usados para calcular e ordenar a importância das features (variáveis de colunas) no modelo. A importância de uma feature é uma medida de quão útil foi essa feature na construção das árvores de decisão no ensemble.
