```python
import pandas as pd
dengue = pd.read_csv('dengbr19.csv')
```

```python
dengue['CLASSI_FIN'].value_counts()
#Possui valor 5, não constava em outras bases
```

<b>.describe()</b> = média, desvio padrão, máximo, mínimo e outras tendências centrais, além da forma da distribuição.

```python
dengue['CLASSI_FIN'] = dengue['CLASSI_FIN'].fillna('0.0')
dengue['CLASSI_FIN'].describe()
```

```python
dengue_sintomas = dengue.iloc[:, 23:44]
#dengue_sintomas.value_counts()
dengue_sintomas.describe()
```

```python
dengue_sintomas.isnull().sum()
#Diferença de quase 300 valores vazios
```

```python
dengue_sintomas = dengue_sintomas.fillna('0')
dengue_sintomas.value_counts()
```

```python
dengue['CLASSI_FIN'] = dengue['CLASSI_FIN'].astype(float)
dengue_sintomas = dengue_sintomas.astype(float)
```

```python
x = dengue_sintomas #variavel preditora - entrada
y = dengue['CLASSI_FIN'][x.index] #variavel target - saida


from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3, random_state=42) 
```

```python
from sklearn.ensemble import ExtraTreesClassifier

modelo = ExtraTreesClassifier()

#Fornecer variáreis para o modelo.
modelo.fit(x_treino, y_treino)
resultado = (modelo.score(x_teste, y_teste)) * 100
print(f'A máquina registrou uma probabilidade de {resultado:.2f}% de acerto.')
```

```python
x_teste[10:40]
```

```python
previsao = modelo.predict(x_teste[10:40])
print(previsao)
#Previu mais valores descartados do que com dengue.
```

<b>Dicionário da classificação baseada nos sintomas:</b>
5 - Descartado |  10 - Dengue | 11 - Dengue com sinais de alarme | 12 - Dengue grave | 13 - Chikungunya | 0.0 Células vazias

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC

modelos = [
    LogisticRegression(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    GradientBoostingClassifier(),
    SVC()
]

for modelo in modelos:
    modelo.fit(x_treino, y_treino)
    resultado = modelo.score(x_teste, y_teste)
    print(f'{modelo.__class__.__name__} score: {resultado}')
```

utilizando <b>SelectKBest</b> baseado na coluna de sintomas

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

f_classif = SelectKBest(score_func = f_classif, k = 5)
fit = f_classif.fit(x, y)
features = fit.transform(x)
print(features) 
```

```python
colunas = fit.get_support(indices=True) #Indentificação das caracterísiticas do fit
dengue_sintomas.iloc[:, colunas] #Localização das colunas
```

<b>Visualização baseados nos dados reais utilizados no treinamento e desempenho do modelo. </b> Aproximação daquilo que se relaciona com X e Y

 Os métodos <b>'model.feature_importances_'</b> e <b>'np.argsort(importances)[::-1]'</b> são usados para calcular e ordenar a importância das features (variáveis de colunas) no modelo. A importância de uma feature é uma medida de quão útil foi essa feature na construção das árvores de decisão no ensemble.

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

previsao = modelo.predict(x_teste)
conf_matrix = confusion_matrix(y_teste, previsao) #matriz de confusão

# Plotar matriz de confusão
plt.figure(figsize=(10,7))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Previsão')
plt.ylabel('Verdadeiro')
plt.title('Matriz de Confusão')
plt.show()
```

Matriz de Confusão serve pra avaliar uma precisão de classificação. Equivale ao número de observações conhecidas por estarem no grupo e prevê-se que esteja em outro grupo. O modelo para precisões deve ser alto.
