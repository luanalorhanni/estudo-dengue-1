```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    'Data': pd.to_datetime(['2023-01-15', '2023-02-20', '2023-03-10', '2023-04-05']),
    'Valor': [10, 20, 15, 25]
}
df = pd.DataFrame(data)
```

```python
# Agrupar por mês e ano e calcular a média (ou qualquer outra métrica desejada)
df_grouped = df.groupby(df['Data'].dt.to_period('M')).mean()
```

```python
# Plotar o gráfico
plt.figure(figsize=(10, 6))
plt.plot(df_grouped.index.astype(str), df_grouped['Valor'], marker='o', linestyle='-')
plt.title('Média de Valores por Mês')
plt.xlabel('Mês e Ano')
plt.ylabel('Valor Médio')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
